```python
#Probability for each predict value
predictValueProbability = {}

for predictValue in df['Class Name'].unique():
    predictValueProbability[predictValue] = len(df[(df['Class Name'] == predictValue)]) / len(df) 

#Probability for each column value    
probabilities = {}

#All unique values that could be predicted
for predictValue in df['Class Name'].unique():
    probabilities[predictValue] = {}
    
    #All columns
    for columnName in df.columns:
        probabilities[predictValue][columnName] = {}
        
        #All unique values a column has
        for columnValue in df[columnName].unique():
            predictValueCount = len(df[(df['Class Name'] == predictValue)])
            predictValueWhichAreColumnValue = len(df[(df['Class Name'] == predictValue) & (df[columnName] == columnValue)])
            probability = predictValueWhichAreColumnValue / predictValueCount
            
            probabilities[predictValue][columnName][columnValue] = probability
```
